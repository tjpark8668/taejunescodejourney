```python
from sklearn.datasets import load_breast_cancer

bc = load_breast_cancer()

print(dir(bc))

# breast cancer를 bc라는 이름으로 줄인 후 bc의 메서드 확인
```

```python
bc.keys()

# bc의 정보 확인
```

```python
bc_data = bc.data

print(bc_data[0])
print(bc_data.shape)

# bc의 data에서 샘플로 첫 번째 행을 가져오면서 shape메서드로 배열의 형상을 확인
```

```python
bc_label = bc.target
print(bc_label.shape)
print(bc.target_names)
print(bc_label)

# bc의 label의 배열과 label의 이름, label을 확인
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(bc_data,
                                                    bc_label,
                                                    test_size = 0.2,
                                                    random_state = 1)

# train_test_split으로 training dataset과 test dataset을 분리
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report

dt = DecisionTreeClassifier(random_state = 1)
dt.fit(X_train, y_train)
y_pred = dt.predict(X_test)

print(classification_report(y_test, y_pred))

# Decision Tree 모델을 학습시키고 예측
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state = 1)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)

print(classification_report(y_test,y_pred))

# Random Forest 모델을 학습시키고 예측
```

```python
from sklearn import svm

svm_model = svm.SVC()
svm_model.fit(X_train, y_train)
y_pred = svm_model.predict(X_test)

print(classification_report(y_test, y_pred))

# svm 모델을 학습시키고 예측
```

```python
from sklearn.linear_model import SGDClassifier

sgd_model = SGDClassifier()
sgd_model.fit(X_train, y_train)
y_pred = sgd_model.predict(X_test)

print(classification_report(y_test, y_pred))

# SGD Classifier 모델을 학습시키고 예측
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)

print(classification_report(y_test, y_pred))

# Logistic Regression 모델을 학습시키고 예측
```

```python
# 위 사례에서는 정확도도 물론 중요하지만, 환자인 사람을 건강하다고 판단하면 큰 문제가 생길 수 있으므로 재현율이 중요하다고 판단, RandomForestClassifier 모델이 적합하다고 판단된다.

rf = RandomForestClassifier(random_state = 1)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)

print(classification_report(y_test,y_pred))
```
